```python
import pandas as pd

# Veri çekme
url = "https://nssdc.gsfc.nasa.gov/planetary/factsheet/"
tablo = pd.read_html(url)

# İlk tabloyu seçelim
df = tablo[0]

# DataFrame'i düzenleme
# İlk sütun gezegen isimleri olacak şekilde ayarlıyoruz, sonra diğer sütunları seçiyoruz
df_cleaned = df.set_index(df.columns[0])  # İlk sütun gezegen isimleri
df_cleaned = df_cleaned.loc[:, :'Mean Temperature']  # Mean Temperature'a kadar olan sütunları seçiyoruz

# Sonuçları kontrol etmek için yazdırma
df_cleaned
```
